```python
#import libraries

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler,PolynomialFeatures
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```python
#download file

file_name = 'https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DA0101EN/coursera/project/kc_house_data_NaN.csv'
df = pd.read_csv(file_name)
```

```python
# view five rows

df.head()
```

```python
#display data types

df.dtypes
```

```python
#statistical summary
df.describe()
```

```python
#data wrangling

#drop columns
df.drop(["id", "Unnamed: 0"],axis = 1,inplace = True)
df.describe()
```

```python
#number of missing values

print("number of NaN values for the column bedrooms :", df['bedrooms'].isnull().sum())
print("number of NaN values for the column bathrooms :", df['bathrooms'].isnull().sum())
```

```python
#replace missing values of the columns with the mean of the column

mean1 = df['bedrooms'].mean()
df['bedrooms'].replace(np.nan , mean1, inplace=True)

mean2 = df['bathrooms'].mean()
df['bathrooms'].replace(np.nan, mean2, inplace=True)

#check
print("number of NaN values for the column bedrooms :", df['bedrooms'].isnull().sum())
print("number of NaN values for the column bathrooms :", df['bathrooms'].isnull().sum())
```

```python
#data analysis
#count the number of house with unique floor values

df['floors'].value_counts()
df['floors'].value_counts().to_frame() #convert to dataframe
```

```python
#boxplot to determine whether houses with a waterfront view or without it have more price outliers

sns.boxplot(x="waterfront", y="price", data=df)
```

```python
#to determine if the feature sqft_above is negatively or positively correlated with price.

sns.regplot(x="sqft_above", y="price", data=df)
plt.ylim(0,)
```

```python
#features correlated with price 

df.corr()['price'].sort_values()
```

```python
#model development

#linear regression 

X = df[['long']]
Y = df['price']
lm = LinearRegression()
lm.fit(X,Y)
lm.score(X, Y)
```

```python
#fit a linear regression model to predict 

X= df[['sqft_living']]
Y= df['price']
lm=LinearRegression()
lm.fit(X,Y)
lm.score(X,Y)
```

```python
#fit linear regression model to predict using multiple features

features =["floors", "waterfront","lat" ,"bedrooms" ,"sqft_basement" ,"view" ,"bathrooms","sqft_living15","sqft_above","grade","sqft_living"]     
X = df[features]
Y = df['price']
lm = LinearRegression()
lm.fit(X, Y)

#calculate R^2 
lm.score(X, Y)
```

```python
#list of tuples created

Input =[('scale',StandardScaler()),
       ('polynomial', PolynomialFeatures(include_bias=False)),
       ('model',LinearRegression())]

#pipeline oject to predict price
pipe = Pipeline(Input)
pipe
pipe.fit(X,Y)
pipe.score(X,Y)
```

```python
#model evaluation and refinement

from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
```

```python
#split the data

features =["floors", "waterfront","lat" ,"bedrooms" ,"sqft_basement" ,"view" ,"bathrooms","sqft_living15","sqft_above","grade","sqft_living"]     
X = df[features]
Y = df['price']

x_train,x_test,y_train,y_test = train_test_split(X,Y, test_size = 0.15,
                                                random_state = 1)

print("the number of test samples: ", x_test.shape[0])
print("number of training samples:",x_train.shape[0])
```

```python
from sklearn.linear_model import Ridge
```

```python
#fit a ridge regression object using training data

RidgeModel = Ridge(alpha=0.1)
RidgeModel.fit(x_train,y_train)
RidgeModel.score(x_test,y_test)
```

```python
#second order polynomial transform on both training and testing data.
#fit a ridge using training data(regularization parameter to 0.1) 
#R^2 score on test data 

pr = PolynomialFeatures(degree=2)
x_train_pr = pr.fit_transform(x_train[features])
x_test_pr = pr.fit_transform(x_test[features])

#ridge model
RidgeModel= Ridge(alpha=0.1)
RidgeModel.fit(x_train_pr, y_train)
RidgeModel.score(x_test_pr, y_test)
```
